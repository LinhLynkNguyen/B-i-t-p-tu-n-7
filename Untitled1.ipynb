```python
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model as linear_model
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
import csv
from sklearn.utils import shuffle
```

```python
x = []
y = []
  
with open('breast_cancer_wisconsin.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
      
    for row in plots:
        x.append(row[0])
        y.append(int(row[1]))
```

```python
df=pd.read_csv('breast_cancer_wisconsin.csv', quoting=csv.QUOTE_NONE)
df
```

```python
x=df.drop(columns=['class'])
y=df['class']
```

```python
def getColor(cancer):
    if cancer == 2:
        return 'red'
    else:
        return 'green'
```

```python
x=np.array(x)
y=np.array(y)
fig, axs=plt.subplots(nrows=1, ncols=2, figsize=(15,7), dpi=90)
for i in range(len(x)):
    if y[i]==2:
        color= 'black'
    else:
        color= 'red'
    axs[0].scatter(x[i][0], x[i][1], color=color)
    axs[1].scatter(x[i][0], x[i][1], color=color)
```

```python
figure, axs = plt.subplots(2,3,figsize=(12,8), dpi=90)
size = 17
axs[0,0].set(xlabel=df.columns[0], ylabel=df.columns[1])
axs[0,0].scatter(x = [sl for sl in df.marg_adhesion], y = [sw for sw in df.single_epith_cell_size], color=getColor('class'))
```

```python

fig, axes=plt.subplots(nrows=1, ncols=3, figsize=(10,5), dpi=100)
for i in range(len(x)):
    if y[i] == 2:
        color ='green'
    else:
        color='red'
    
    axes[0].scatter(x['marg_adhesion'][i], x['single_epith_cell_size'][i], color=color,)
    axes[1].scatter(x['single_epith_cell_size'][i], x['bare_nuclei'][i], color=color)
    axes[2].scatter(x['marg_adhesion'][i], x['bare_nuclei'][i], color=color)
```

```python
# V  th 3D
fig = plt.figure(dpi=200)
ax1 = fig.add_subplot(121, projection='3d') # nrows, ncols, index default: (1, 1, 1)

for i in df.index:
    color = getColor(df['class'][i])
    x = df.marg_adhesion[i]
    y = df.single_epith_cell_size[i]
    z = df['class'][i]
    ax1.scatter(x, y, z, c=color, s=7)
```
